```python
from __future__ import print_function
__author__= 'ZYJ'
import os 

from scipy.optimize import curve_fit, minimize
import pylab as pl
import numpy as np
from scipy.stats import norm
import matplotlib.mlab as mlab
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
%pylab inline
```

##Poisson Distribution

```python
#distributions = ['pois', 'binomial', 'chisq', 'normal', 'uniform', 'exponential']
distributions = ['pois']
```

```python
mymean=100
df=100
mysize=2000/(np.array(range(1,100)+[10]))
md={}
means={}
```

```python
numpy.random.seed(80)
```

```python
md['pois']={}
for n in mysize:
    md['pois'][n] = np.random.poisson(df, size=n)

means['pois'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['pois'].iterkeys():
    means['pois'][nn] = md['pois'][nn].mean()
    axchisq_mu_n.plot(nn, means['pois'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
```

```python
allmeans=[]
for n in means['pois']:
    allmeans.append(means['pois'][n])
pl.figure(figsize=(10,10))
pl.hist(allmeans,bins=30)
pl.xlabel('sample mean')
pl.ylabel('N')
```

For these six distribution,we will expect the sample mean is closer to the population mean when sample size is larger enough.

```python
allmeans=[]
for n in means['pois']:
    allmeans.append(means['pois'][n])
for n in means['binomial']:
    allmeans.append(means['binomial'][n])
for n in means['chisq']:
    allmeans.append(means['chisq'][n])
for n in means['normal']:
    allmeans.append(means['normal'][n])
for n in means['uniform']:
    allmeans.append(means['uniform'][n])
for n in means['exponential']:
    allmeans.append(means['exponential'][n])
pl.figure(figsize=(20,20))
pl.hist(allmeans,bins=80)
pl.xlabel('sample mean')
pl.ylabel('N')
```

```python
from scipy.stats import norm
import matplotlib.mlab as mlab

(mu, sigma) = norm.fit(allmeans)
pl.figure(figsize=(20,20))
n, bins, patches = plt.hist(allmeans, normed=1, bins = 80,  facecolor='blue', alpha=0.5)
y = mlab.normpdf( bins, mu, sigma)
l = pl.plot(bins, y, 'r--', linewidth=2)
```
